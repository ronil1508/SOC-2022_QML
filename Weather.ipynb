```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset= pd.read_csv('/weatherHistory.csv')
print(dataset.shape)
```

```python
print(dataset.describe())
```

```python
dataset.head()
```

```python
dataset1 = dataset[['Temperature (C)','Humidity']]
dataset1.head()
```

```python
X = np.array(dataset1['Temperature (C)']).reshape(-1,1)
y = np.array(dataset1['Humidity']).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

We can clearly see that linear regression underfits the triaining data

Now we perform multi-variable linear regression


```python
dataset2 = dataset[['Temperature (C)','Humidity','Pressure (millibars)']]
dataset2.head()
```

```python
X = np.array(dataset2[['Temperature (C)','Pressure (millibars)']])
y = np.array(dataset2['Humidity'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
lr.fit(X_train,y_train)
print(lr.score(X_train,y_train))
print(lr.score(X_test,y_test))
```

Multi variable linear regression also underfits the data.
